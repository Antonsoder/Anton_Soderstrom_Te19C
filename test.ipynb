```python
import math
import random as rnd
import matplotlib.pyplot as plt

forsok = float(input("Hur många gånger kastar du?"))

loop = 0
while loop < forsok:
    loop = loop + 1
    y = rnd.uniform(-1, 1)
    x = rnd.uniform(-1, 1)
    x2 = x * x
    y2 = y * y
    c = x2 + y2
    hypotentusan = math.sqrt(c)

    if hypotentusan > 1:
        plt.plot(y, x, '*r')
    else:
         plt.plot(y, x, '*g')

plt.title("Pricktavla")
plt.show()
```
