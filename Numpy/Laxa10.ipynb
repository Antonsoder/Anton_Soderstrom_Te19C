```python
#Uppgift 3
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,11)

def f(k,x,m):
    y = k*x + m
    return y
print(y)


plt.plot(x,f(3,x,1))
plt.title("En rät linje")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
