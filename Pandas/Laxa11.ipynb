```python
#Uppgift 1a

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sysselsattning.csv")

df = df["2019"][0:12]

förvärsarbetande = df.sum().sum()

print(f"Under 2019 var det {förvärsarbetande} st förvärsarbetande mellan åldrarna 16-44")

#Uppgift 1b

df = pd.read_csv("sysselsattning.csv")

df = df["2019"][13:26]

ejförvärsarbetande = df.sum().sum()

print(f"Under 2019 var det {ejförvärsarbetande} st som inte var förvärsarbetande mellan åldrarna 16-44")

#Uppgift 1c

statistik =  [förvärsarbetande, ejförvärsarbetande]

labels = ["förvärsarbetande", "ejförvärsarbetande"]



plt.pie(statistik, labels = labels, autopct ="")
plt.title("Antal förvärsarbetande / inte förvärsarbetande")
plt.show()

```

```python
#Uppgift 2a
import pandas as pnd

df = pd.read_csv("sysselsattning.csv")

df['förvärsarbetande'] = df['2019'][:12] 

df['män'] = df['förvärsarbetande'][::2]

antal_män = df['män'].sum()

print(f"{antal_män} män förvärsarbetade under 2019 mellan åldrarna 16-44")

#Uppgift 2b

df['förvärsarbetande'] = df['2019'][:12] 

df['kvinnor'] = df['förvärsarbetande'][1::2]

antal_kvinnor = df['kvinnor'].sum()

print(f"{antal_kvinnor} kvinnor förvärsarbetade under 2019 mellan åldrarna 16-44")

#Uppgift 2c
df['ejförvärsarbetande'] = df['2019'][12:25] 

df['män'] = df['ejförvärsarbetande'][::2]

ejarbetande_män = df['män'].sum()

print(f"Antal ej förvärsarbetande män under året 2019 mellan åldrarna 16-44 var: {ejarbetande_män}st")

#Uppgift 2d

df['ejförvärsarbetande'] = df['2019'][12:25] 

df['kvinnor'] = df['ejförvärsarbetande'][1::2]

ejarbetande_kvinnor = df['kvinnor'].sum()

print(f"Antal ej förvärsarbetande kvinnor under året 2019 mellan åldrarna 16-44 var: {ejarbetande_kvinnor}st")

#Uppgift 2e

import matplotlib.pyplot as plt

arbete = ['Män (ja)', 'Kvinnor (ja)', 'Män (nej)', 'Kvinnor (nej)']

antal_personer = [antal_män, antal_kvinnor, ejarbetande_män, ejarbetande_kvinnor]

plt.bar(arbete , antal_personer, width = 0.5)


plt.title('Förvärsarbetare/ejförvärsarbetare')
plt.ylabel('Antal (Miljoner)')
plt.show()



```
