```python
import pandas as pd
import numpy as np

df = pd.read_csv("Matteprov.csv")

#print(df.head(3))
#print(df.tail(2))
#df = df.dropna() #Slänger raderna som inehåller nan
df.fillna(0) #Lägger till nollor
df["Total"] = df.sum(axis=1)
df["Betyg"] = np.where(df["Total"] > 7, "Godkänt", "Underkänt")
df
```

```python
#PLocka ut kolumner emd kolumnamn

df["Betyg"]
df["Elev"]
```

```python
#Conditional Filtering

df_G = df[df["Betyg"]=="Godkänt"] #True = godkänt
df_U = df[df["Betyg"]=="Underkänt"] #True = underkänd
df_U
```

```python
# Data visualisering

import matplotlib.pyplot as plt

betygslista = df["Betyg"].value_counts().tolist() # Räknar antalet godkända samt underkända och lägger i en lista
labels = df["Betyg"].unique() # plockar ut våra labels, dsvs udnerkänt och godkänt

plt.bar(labels, betygslista, width=.4, color = ["green", "red"])
plt.title("Antal godkända/underkända i klassen")
plt.show()
```
